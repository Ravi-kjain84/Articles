```python
import pandas as pd
import numpy as np

df = pd.read_excel("/content/drive/MyDrive/Colab data/z_score_data.xlsx",sheet_name = 'Raw_data')

display(df.head(2))
# Group by bank and calculate ROA, ETA for each bank per year
df['ROA'] = df['Net Income'] / df['Total Assets']
df['ETA'] = df['Total Equity'] / df['Total Assets']

# Calculate mean ROA, mean ETA and standard deviation of ROA by bank
bank_stats = df.groupby('Bank').agg(
    Mean_ROA=('ROA', 'mean'),
    Mean_ETA=('ETA', 'mean'),
    Std_ROA=('ROA', 'std')
)

# Calculate the Z-score for each bank
bank_stats['Z-score'] = (bank_stats['Mean_ROA'] + bank_stats['Mean_ETA']) / bank_stats['Std_ROA']

display(bank_stats[['Mean_ROA', 'Mean_ETA', 'Std_ROA', 'Z-score']])
```
